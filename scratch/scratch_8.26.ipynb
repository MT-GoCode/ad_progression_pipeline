```python
import pandas as pd
from sklearn.impute import KNNImputer


def train_knn_imputers(df: pd.DataFrame, numerical_feats: list, n_neighbors: int = 50) -> dict:
    imputer_map = {}

    for feat in numerical_feats:
        # Create a KNNImputer for the specific feature
        imputer = KNNImputer(n_neighbors=n_neighbors)

        # Fit the imputer on the DataFrame using only the specific feature
        df_feat = df[[feat]]
        imputer.fit(df_feat)

        # Store the trained imputer in the map with the feature name as the key
        imputer_map[feat] = imputer

    return imputer_map


import numpy as np

# Train imputers
do = [_ for _ in NUMERICAL_FEATURES if _ in input_.columns]
imputer_map = train_knn_imputers(input_, do)

# Test data to impute (we'll use the same test data for each feature)
test_data = np.array([[np.nan]])

# Check the imputation results for each feature
for feat in do:
    imputer = imputer_map[feat]
    imputed_value = imputer.transform(test_data)
    print(f"Imputed value for {feat}: {imputed_value[0][0]}")
```
